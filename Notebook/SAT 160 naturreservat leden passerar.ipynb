```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Exporterar (med prefix SAT_160_):
  geojson/SAT_160_reservat.geojson
  geojson/SAT_160_etapper.geojson
  geojson/SAT_160_etapp_i_skydd.geojson
  stats/SAT_160_etapp_andel_skydd.csv

K√§llor:
- Wikidata SPARQL: Q131318799 (SAT), P3018 (reservat), P402 (OSM relation), P10689 (OSM way)
- Overpass API: h√§mtar OSM-geometrier och konverterar till GeoJSON via osm2geojson
"""

import os
import json
import time
import requests
import pandas as pd
from pathlib import Path
from shapely.geometry import shape, LineString, MultiLineString, Polygon, MultiPolygon, mapping
from shapely.ops import unary_union, linemerge
from pyproj import Transformer
from SPARQLWrapper import SPARQLWrapper, JSON as SPARQL_JSON

try:
    from osm2geojson import json2geojson  # pip install osm2geojson
except Exception as e:
    raise SystemExit("‚ùå Paketet 'osm2geojson' saknas. Installera med: pip install osm2geojson") from e

# ---------- Konfig ----------
ISSUE_PREFIX = os.environ.get("ISSUE_PREFIX", "SAT_160_")  # kan √§ndras via env
WDQS = "https://query.wikidata.org/sparql"
OVERPASS = "https://overpass-api.de/api/interpreter"

OUT_DIR = Path("geojson").resolve()
STATS_DIR = Path("stats").resolve()
OUT_DIR.mkdir(exist_ok=True, parents=True)
STATS_DIR.mkdir(exist_ok=True, parents=True)

F_RESERVAT = OUT_DIR / f"{ISSUE_PREFIX}reservat.geojson"
F_ETAPPER = OUT_DIR / f"{ISSUE_PREFIX}etapper.geojson"
F_ETAPP_I_SKYDD = OUT_DIR / f"{ISSUE_PREFIX}etapp_i_skydd.geojson"
F_STATS = STATS_DIR / f"{ISSUE_PREFIX}etapp_andel_skydd.csv"

print(f"üìÅ GEOJSON: {OUT_DIR}")
print(f"üìÅ STATS  : {STATS_DIR}")
print(f"‚Ü™Ô∏è Prefix : {ISSUE_PREFIX}")

# ---------- Hj√§lpare ----------
def sparql_query(q: str) -> dict:
    s = SPARQLWrapper(WDQS, agent="SAT-Exporter/1.1 (https://github.com/salgo60)")
    s.setQuery(q)
    s.setReturnFormat(SPARQL_JSON)
    return s.query().convert()

def overpass_fetch(elements: list[str]) -> dict:
    """elements: lista av 'relation/123' eller 'way/456'"""
    if not elements:
        return {"elements": []}
    ids_by_type = {"relation": [], "way": []}
    for e in elements:
        t, i = e.split("/")
        ids_by_type[t].append(i)

    parts = []
    if ids_by_type["relation"]:
        parts.append(f"relation(id:{','.join(ids_by_type['relation'])});")
    if ids_by_type["way"]:
        parts.append(f"way(id:{','.join(ids_by_type['way'])});")

    q = f"""
    [out:json][timeout:180];
    (
      {' '.join(parts)}
    );
    out body geom;
    >;
    out body geom;
    """
    r = requests.post(OVERPASS, data={"data": q}, timeout=180)
    r.raise_for_status()
    return r.json()

def overpass_to_geojson(overpass_json: dict) -> dict:
    gj = json2geojson(overpass_json)
    for f in gj["features"]:
        f.setdefault("properties", {})
        if "id" not in f["properties"] and "id" in f:
            f["properties"]["id"] = f["id"]
    return gj

def reproject_length_m(geom, src_epsg="EPSG:4326", dst_epsg="EPSG:3006"):
    transformer = Transformer.from_crs(src_epsg, dst_epsg, always_xy=True)
    def _reproj_coords(coords):
        for x, y in coords:
            X, Y = transformer.transform(x, y)
            yield (X, Y)
    if isinstance(geom, LineString):
        return LineString(list(_reproj_coords(list(geom.coords)))).length
    if isinstance(geom, MultiLineString):
        parts = [LineString(list(_reproj_coords(list(ls.coords)))) for ls in geom.geoms]
        return MultiLineString(parts).length
    return 0.0

def ensure_mls(geom):
    if isinstance(geom, LineString):
        return MultiLineString([geom])
    if isinstance(geom, MultiLineString):
        return geom
    raise TypeError("Expected LineString or MultiLineString")

def ensure_mpoly(geom):
    if isinstance(geom, Polygon):
        return MultiPolygon([geom])
    if isinstance(geom, MultiPolygon):
        return geom
    raise TypeError("Expected Polygon or MultiPolygon")

def write_fc(features, path: Path):
    fc = {"type": "FeatureCollection", "features": features}
    with open(path, "w", encoding="utf-8") as f:
        json.dump(fc, f, ensure_ascii=False)
    print(f"üíæ Skrev {path} ({len(features)} features)")

# ---------- 1) H√§mta ID:n fr√•n Wikidata ----------
SPARQL = """
SELECT DISTINCT ?item ?itemLabel ?typ ?osmid ?osmidWay ?etappLabel WHERE {
  {
    # Naturreservat kopplade till SAT
    wd:Q131318799 wdt:P3018 ?item .
    BIND("Naturreservat"@sv AS ?typ)
    OPTIONAL { ?item wdt:P402 ?osmid. }      # relation
    OPTIONAL { ?item wdt:P10689 ?osmidWay. } # way (OSM way)
  }
  UNION
  {
    # Etapper (delstr√§ckor)
    ?item wdt:P361 wd:Q131318799 .
    BIND("Etapp"@sv AS ?typ)
    OPTIONAL { ?item wdt:P402 ?osmid. }      # relation (f√∂r etapp)
    OPTIONAL { ?item rdfs:label ?etappLabel. FILTER(LANG(?etappLabel)='sv') }
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "sv,en". }
}
"""

print("üîé H√§mtar objekt fr√•n Wikidata ‚Ä¶")
rows = sparql_query(SPARQL)["results"]["bindings"]

reservat_ids, etapp_rel_ids = [], []
name_by_id = {}

for r in rows:
    typ = r["typ"]["value"]
    itemLabel = r.get("itemLabel", {}).get("value", "")
    if typ == "Naturreservat":
        if "osmid" in r:
            rid = r["osmid"]["value"]
            reservat_ids.append(f"relation/{rid}")
            name_by_id[f"relation/{rid}"] = itemLabel
        if "osmidWay" in r:
            wid = r["osmidWay"]["value"]
            reservat_ids.append(f"way/{wid}")
            name_by_id[f"way/{wid}"] = itemLabel
    else:  # Etapp
        if "osmid" in r:
            rid = r["osmid"]["value"]
            etapp_rel_ids.append(f"relation/{rid}")
            etapp_label = r.get("etappLabel", {}).get("value", itemLabel)
            name_by_id[f"relation/{rid}"] = etapp_label

reservat_ids = sorted(set(reservat_ids))
etapp_rel_ids = sorted(set(etapp_rel_ids))
print(f"‚Ä¢ Reservat OSM-id: {len(reservat_ids)}")
print(f"‚Ä¢ Etapp-relationer: {len(etapp_rel_ids)}")

# ---------- 2) H√§mta geometrier fr√•n Overpass ----------
def fetch_layer(elements, sleep=2.0):
    if not elements:
        return {"type": "FeatureCollection", "features": []}
    BATCH = 120
    features = []
    for i in range(0, len(elements), BATCH):
        chunk = elements[i:i+BATCH]
        j = overpass_fetch(chunk)
        gj = overpass_to_geojson(j)
        for f in gj["features"]:
            osm_key = f"{f['properties'].get('type','')}/{f['properties'].get('id','')}".strip("/")
            if osm_key in name_by_id:
                f["properties"]["name_from_wd"] = name_by_id[osm_key]
            features.append(f)
        time.sleep(sleep)
    return {"type": "FeatureCollection", "features": features}

print("‚¨áÔ∏è H√§mtar naturreservat ‚Ä¶")
gj_reservat_raw = fetch_layer(reservat_ids)
print("‚¨áÔ∏è H√§mtar etapper ‚Ä¶")
gj_etapp_raw = fetch_layer(etapp_rel_ids)

# ---------- 3) Rensa & sl√• ihop ----------
def collect_polygons_from_features(fc):
    out = []
    for f in fc["features"]:
        geom = f.get("geometry")
        if not geom:
            continue
        g = shape(geom)
        props = dict(f.get("properties", {}))
        name = props.get("name") or props.get("name_from_wd") or props.get("ref") or ""
        if isinstance(g, (Polygon, MultiPolygon)):
            out.append({"geometry": g, "properties": {"name": name} | props})
        elif isinstance(g, LineString) and g.is_ring:
            out.append({"geometry": Polygon(g), "properties": {"name": name} | props})
        elif isinstance(g, MultiLineString):
            merged = linemerge(g)
            if isinstance(merged, LineString) and merged.is_ring:
                out.append({"geometry": Polygon(merged), "properties": {"name": name} | props})
    return out

def collect_lines_from_features(fc):
    out = []
    for f in fc["features"]:
        geom = f.get("geometry")
        if not geom:
            continue
        g = shape(geom)
        props = dict(f.get("properties", {}))
        name = props.get("name_from_wd") or props.get("name") or props.get("ref") or ""
        if isinstance(g, (LineString, MultiLineString)):
            out.append({"geometry": g, "properties": {"etapp": name} | props})
        elif isinstance(g, Polygon):
            out.append({"geometry": g.boundary, "properties": {"etapp": name} | props})
    return out

res_polys = collect_polygons_from_features(gj_reservat_raw)
etapp_lines = collect_lines_from_features(gj_etapp_raw)

def dissolve_by_name(items, name_key):
    groups = {}
    for it in items:
        nm = it["properties"].get(name_key) or "ok√§nt"
        groups.setdefault(nm, []).append(it["geometry"])
    dissolved = []
    for nm, geoms in groups.items():
        u = unary_union(geoms)
        dissolved.append((nm, u))
    return dissolved

res_by_name = dissolve_by_name(res_polys, "name")
etapp_by_name = dissolve_by_name(etapp_lines, "etapp")

# ---------- 4) Skriv GeoJSON ----------
features_res = []
for nm, g in res_by_name:
    try:
        mpoly = ensure_mpoly(g)
        features_res.append({
            "type": "Feature",
            "properties": {"name": nm, "layer": "Naturreservat"},
            "geometry": mapping(mpoly)
        })
    except Exception:
        continue
write_fc(features_res, F_RESERVAT)

features_etapp = []
for nm, g in etapp_by_name:
    try:
        mls = ensure_mls(g)
        features_etapp.append({
            "type": "Feature",
            "properties": {"etapp": nm, "layer": "Etapp"},
            "geometry": mapping(mls)
        })
    except Exception:
        continue
write_fc(features_etapp, F_ETAPPER)

# ---------- 5) Andel per etapp i skydd ----------
reserve_union = unary_union([shape(f["geometry"]) for f in features_res]) if features_res else None

stats_rows = []
features_in_protected = []

if reserve_union and not reserve_union.is_empty:
    for f in features_etapp:
        etapp_name = f["properties"]["etapp"]
        g = shape(f["geometry"])
        inter = g.intersection(reserve_union)
        tot_m = reproject_length_m(g)
        in_m = reproject_length_m(inter) if not inter.is_empty else 0.0
        pct = (in_m / tot_m * 100.0) if tot_m > 0 else 0.0
        stats_rows.append({
            "etapp": etapp_name,
            "langd_tot_m": round(tot_m, 1),
            "langd_i_skydd_m": round(in_m, 1),
            "andel_i_skydd_procent": round(pct, 2),
        })
        if not inter.is_empty:
            if isinstance(inter, Polygon):
                inter = inter.boundary
            if isinstance(inter, MultiPolygon):
                inter = unary_union([p.boundary for p in inter.geoms])
            if isinstance(inter, LineString):
                inter = MultiLineString([inter])
            features_in_protected.append({
                "type": "Feature",
                "properties": {
                    "etapp": etapp_name,
                    "layer": "Etapp i skydd",
                    "andel_i_skydd_procent": round(pct, 2)
                },
                "geometry": mapping(inter)
            })
else:
    print("‚ö†Ô∏è Hittade inga reservatsytor att j√§mf√∂ra mot (tom union).")

write_fc(features_in_protected, F_ETAPP_I_SKYDD)

df = pd.DataFrame(stats_rows).sort_values("etapp")
df.to_csv(F_STATS, index=False, encoding="utf-8")
print(f"üìä Skrev {F_STATS}")

print("\n‚úÖ Klart. Importera i uMap i denna ordning:")
print(f"  1) {F_RESERVAT}")
print(f"  2) {F_ETAPPER}")
print(f"  3) {F_ETAPP_I_SKYDD}")
print(f"   ‚ÑπÔ∏è Statistik: {F_STATS}")
```
