```python
import requests

# URL till JSON-filen
url = "https://map.stockholmarchipelagotrail.com/data/vessels-identity.json"

# Ersätt med rätt property när den finns
PROPERTY = "P14545"

# Hämta JSON
response = requests.get(url)
response.raise_for_status()
data = response.json()

# Skapa QuickStatements
lines = []

for obj in data.values():
    wikidata = obj.get("wikidata")
    satid = obj.get("satId")

    if wikidata and satid:
        lines.append(f'{wikidata}\t{PROPERTY}\t"{satid}"')

# Spara fil
output_file = "vessels_quickstatements.txt"

with open(output_file, "w", encoding="utf-8") as f:
    f.write("\n".join(lines))

print(f"Skrev {len(lines)} QuickStatements till {output_file}")

# Visa 100 första raderna
for line in lines[:100]:
    print(line)
```
