```python
import time
import datetime  
start_time = time.time()
start_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
print(f"Started: {start_str}")
```

```python
import requests

url = "https://map.stockholmarchipelagotrail.com/data/geojson/poi-concordance.json"
data = requests.get(url).json()

wikidata_ids = [
    k.replace("wikidata:", "")
    for k in data["satIdOf"].keys()
    if k.startswith("wikidata:")
]

print(f"{len(wikidata_ids)} Wikidata-objekt")
print(wikidata_ids[:10])
```

```python
import requests
import time
import json

CONCORDANCE_URL = "https://map.stockholmarchipelagotrail.com/data/geojson/poi-concordance.json"

print("Loading concordance...")
concordance = requests.get(CONCORDANCE_URL).json()

sat_of = concordance["satIdOf"]

# --------------------------------------------------
# Build indexes
# --------------------------------------------------

wikidata_to_sat = {}
sat_to_osm = {}

for key, sat_id in sat_of.items():

    if key.startswith("wikidata:"):
        qid = key.replace("wikidata:", "")
        wikidata_to_sat[qid] = sat_id

    elif key.startswith("osm:"):
        sat_to_osm.setdefault(sat_id, []).append(key)

print(f"Wikidata objects: {len(wikidata_to_sat)}")

# --------------------------------------------------
# Check Wikidata
# --------------------------------------------------

alive = []
redirects = []
deleted = []
errors = []

for i, qid in enumerate(sorted(wikidata_to_sat.keys()), start=1):

    print(f"[{i}/{len(wikidata_to_sat)}] {qid}", end="\r")

    try:

        url = (
            f"https://www.wikidata.org/wiki/"
            f"Special:EntityData/{qid}.json"
        )

        r = requests.get(
            url,
            timeout=30,
            headers={
                "User-Agent":
                "SAT Wikidata Quality Check/1.0"
            }
        )

        if r.status_code == 404:

            deleted.append({
                "qid": qid,
                "sat_id": wikidata_to_sat[qid],
                "osm": sat_to_osm.get(
                    wikidata_to_sat[qid],
                    []
                )
            })

            continue

        if r.status_code != 200:

            errors.append({
                "qid": qid,
                "status_code": r.status_code
            })

            continue

        entities = r.json()["entities"]

        returned_qid = next(iter(entities.keys()))

        if returned_qid != qid:

            redirects.append({
                "qid": qid,
                "target": returned_qid,
                "sat_id": wikidata_to_sat[qid],
                "osm": sat_to_osm.get(
                    wikidata_to_sat[qid],
                    []
                )
            })

        else:
            alive.append(qid)

    except Exception as e:

        errors.append({
            "qid": qid,
            "error": str(e)
        })

    time.sleep(0.1)

print()
print("=" * 60)
print("SUMMARY")
print("=" * 60)
print(f"Alive     : {len(alive)}")
print(f"Redirects : {len(redirects)}")
print(f"Deleted   : {len(deleted)}")
print(f"Errors    : {len(errors)}")

# --------------------------------------------------
# Redirects
# --------------------------------------------------

if redirects:

    print()
    print("=" * 60)
    print("REDIRECTS")
    print("=" * 60)

    for item in redirects:

        print()
        print(
            f"{item['qid']} -> {item['target']}"
        )
        print(
            f"SAT: {item['sat_id']}"
        )

        for osm in item["osm"]:

            _, osm_type, osm_id = osm.split(":")

            print(
                f"OSM: https://www.openstreetmap.org/"
                f"{osm_type}/{osm_id}"
            )

# --------------------------------------------------
# Deleted
# --------------------------------------------------

if deleted:

    print()
    print("=" * 60)
    print("DELETED WIKIDATA OBJECTS")
    print("=" * 60)

    for item in deleted:

        print()
        print(
            f"Wikidata: {item['qid']}"
        )
        print(
            f"SAT: {item['sat_id']}"
        )

        if item["osm"]:

            print("OSM objects:")

            for osm in item["osm"]:

                _, osm_type, osm_id = osm.split(":")

                print(
                    f"  https://www.openstreetmap.org/"
                    f"{osm_type}/{osm_id}"
                )

        else:
            print(
                "No linked OSM object found"
            )

# --------------------------------------------------
# Errors
# --------------------------------------------------

if errors:

    print()
    print("=" * 60)
    print("ERRORS")
    print("=" * 60)

    for item in errors:
        print(item)

# --------------------------------------------------
# JSON report
# --------------------------------------------------

report = {
    "alive": len(alive),
    "redirects": redirects,
    "deleted": deleted,
    "errors": errors,
}

with open(
    "wikidata_quality_report.json",
    "w",
    encoding="utf-8"
) as f:
    json.dump(
        report,
        f,
        indent=2,
        ensure_ascii=False
    )

print()
print(
    "Saved: wikidata_quality_report.json"
)

```
