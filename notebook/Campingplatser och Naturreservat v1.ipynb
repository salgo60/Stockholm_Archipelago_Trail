```python
import time
from datetime import datetime

start_time = time.time()
print("Start:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```

```python
import folium
from folium import FeatureGroup, Icon, Popup, LayerControl
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
from IPython.display import display

# SPARQL query to get SAT reservat + tältplatser with type
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
sparql.setReturnFormat(JSON)
sparql.setQuery("""
SELECT DISTINCT ?item ?itemLabel ?typ ?SATLabel ?coord ?img ?www ?Naturvårdsreg ?Facebook ?Instagram ?InstagramL ?Googlemap WHERE {
  {
    wd:Q131318799 wdt:P3018 ?item.
    BIND("Naturreservat"@sv AS ?typ)
    OPTIONAL { ?item wdt:P625 ?coord. }
    OPTIONAL { ?item wdt:P18 ?img. }
    OPTIONAL { ?item wdt:P856 ?www. }
    OPTIONAL {
      ?item wdt:P2789 ?SATid.
      ?SATid wdt:P361 wd:Q131318799.
      ?SATid rdfs:label ?SATLabel. FILTER(LANG(?SATLabel)="sv")
    }
    OPTIONAL { ?item wdt:P3613 ?Naturvårdsregid. }
    BIND(IF(BOUND(?Naturvårdsregid), URI(CONCAT("https://skyddadnatur.naturvardsverket.se/sknat/?nvrid=", STR(?Naturvårdsregid))), "") AS ?Naturvårdsreg)
  }
  UNION
  {
    VALUES ?sleep { wd:Q832778 }
    ?item wdt:P6104 wd:Q134294510;
          wdt:P625 ?coord;
          wdt:P31 ?sleep.
    BIND("Tältplats"@sv AS ?typ)
    OPTIONAL {
      ?item wdt:P2789 ?SATid.
      ?SATid wdt:P361 wd:Q131318799.
      ?SATid rdfs:label ?SATLabel. FILTER(LANG(?SATLabel)="sv")
    }
    OPTIONAL {?item wdt:P856 ?www}
    OPTIONAL {?item wdt:P2003 ?InstaU}
    OPTIONAL {?item wdt:P4173 ?InstaL}
    OPTIONAL {?item wdt:P2013 ?FB}
    OPTIONAL {?item wdt:P3749 ?Gmap}
    OPTIONAL {?item wdt:P18 ?img. }
    BIND(URI(CONCAT("https://www.facebook.com/", ?FB)) AS ?Facebook)
    BIND(URI(CONCAT("https://www.instagram.com/", ?InstaU)) AS ?Instagram)
    BIND(URI(CONCAT("https://www.instagram.com/explore/locations/", ?InstaL)) AS ?InstagramL)
    BIND(URI(CONCAT("https://www.google.com/maps?cid=", ?Gmap)) AS ?Googlemap)
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "sv,en". }
}
""")
results = sparql.query().convert()

# Parse results into DataFrame
rows = []
for r in results["results"]["bindings"]:
    def get(field): return r[field]["value"] if field in r else None
    coord = get("coord")
    lat, lon = None, None
    if coord:
        coord = coord.replace("Point(", "").replace(")", "")
        lon, lat = map(float, coord.split())
    rows.append({
        "label": get("itemLabel"),
        "typ": get("typ"),
        "lat": lat,
        "lon": lon,
        "sat": get("SATLabel"),
        "img": get("img"),
        "www": get("www"),
        "Naturvårdsreg": get("Naturvårdsreg"),
        "Facebook": get("Facebook"),
        "Instagram": get("Instagram"),
        "InstagramL": get("InstagramL"),
        "Googlemap": get("Googlemap")
    })
df = pd.DataFrame(rows).dropna(subset=["lat", "lon"])

# Create Folium map
m = folium.Map(location=[59.4, 18.8], zoom_start=8, tiles="OpenStreetMap")

# Layer groups
layers = {}
colors = {"Naturreservat": "green", "Tältplats": "orange"}

for _, row in df.iterrows():
    layer_key = row["typ"]
    if layer_key not in layers:
        layers[layer_key] = FeatureGroup(name=layer_key)
        m.add_child(layers[layer_key])

    links = []
    if row["www"]: links.append(f'<a href="{row["www"]}" target="_blank">Webbplats</a>')
    if row["Naturvårdsreg"]: links.append(f'<a href="{row["Naturvårdsreg"]}" target="_blank">Naturvårdsregistret</a>')
    if row["Facebook"]: links.append(f'<a href="{row["Facebook"]}" target="_blank">Facebook</a>')
    if row["Instagram"]: links.append(f'<a href="{row["Instagram"]}" target="_blank">Instagram</a>')
    if row["InstagramL"]: links.append(f'<a href="{row["InstagramL"]}" target="_blank">Instagram-plats</a>')
    if row["Googlemap"]: links.append(f'<a href="{row["Googlemap"]}" target="_blank">Google Maps</a>')

    html = f"<b>{row['label']}</b><br>"
    if row["sat"]: html += f"<i>{row['sat']}</i><br>"
    if row["img"]: html += f'<img src="{row["img"]}" width="150"><br>'
    html += "<br>".join(links)

    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=Popup(html, max_width=250),
        icon=Icon(color=colors.get(row["typ"], "blue"), icon="info-sign"),
    ).add_to(layers[layer_key])

# Add layer control
m.add_child(LayerControl())
display(m)
```

```python
 # End timer and calculate duration
end_time = time.time()
elapsed_time = end_time - start_time

# Print current date and total time
print("Date:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print("Total time elapsed: {:.2f} seconds".format(elapsed_time))
```
