```python
import geopandas as gpd
from shapely.geometry import box
import folium

# Define bounding boxes
data = [
    {"qid": "Q123456", "name": "Furusund", "geometry": box(18.8, 59.74, 18.86, 59.78)},
    {"qid": "Q234567", "name": "Grinda", "geometry": box(18.58, 59.40, 18.68, 59.46)},
]

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(data, crs="EPSG:4326")

# Center map
center = gdf.geometry.unary_union.centroid
m = folium.Map(location=[center.y, center.x], zoom_start=11)

# Draw bounding boxes
for _, row in gdf.iterrows():
    coords = [(lat, lon) for lon, lat in row.geometry.exterior.coords]
    folium.Polygon(
        locations=coords,
        color="blue",
        fill=True,
        fill_opacity=0.3,
        tooltip=row["name"]
    ).add_to(m)

m
```
